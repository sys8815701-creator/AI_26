```python
# 머신러닝 알고리즘은 지도 학습, 비지도 학습으로 나누어진다.
# 지도 학습 : 훈련하기 위한 데이터와 정답이 필요하다.
# 지도 학습에서 데이터를 입력(input), 정답을 타깃(target) 이라고 한다. → 둘을 합친 것을 훈련 데이터(training data)라고 한다.
# 특성(feature) : 입력으로 사용된 길이와 무게

# MBC 마켓에서 살아있는 생선을 최초로 판매하고자 한다.
# 고객이 온라인 주문을 하면 가장 빠른 물류 센터에서 신선한 생선을 곧바로 배송한다.
# 이때 발생할 수 있는 문제는 생선을 포장하는 직원이 생선의 종류를 알지 못한다는 것이다. → IP 카메라와 객체 추적 기술을 활용한다.

# 우선 도미와 빙어 2마리의 객체를 찾도록 훈련을 시작한다.
# 1. 전처리 데이터가 필요하다. → 도마 생선의 길이, 무게를 학습시켜 보자. (길이와 무게라는 특성 활용)
```

```python
# 도미 생선의 길이 : bream_length 도미용 길이 리스트
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]

# 도미 생선의 무게 : bream_weigh 도미용 무게 리스트
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
# 특성 : 데이터의 특징
# 첫 번째 도미의 특성 : 길이와 무게 (25.4cm, 242.0g)
# 두 번째 도미의 특성 : 길이와 무게 (26.3cm, 290.0g)

# 두 개의 특성을 숫자로 보는 것보다 그래프로 표현하면 데이터를 더 잘 이해할 수 있다.
# 길이는 x축, 무게는 y축 → 산점도 (두 좌표의 관계)
```

```python
import matplotlib.pylab as plt # matplotlib : 파이썬에서 과학 계산용 그래프를 그리는 대표적인 패키지
# https://wikidocs.net/book/5011: matplotlib 사용 방법

plt.scatter(bream_length, bream_weight) # 차트에 값을 넣는다.
# 차트 종류 (x축 : bream_length, y축 : bream_weight)
plt.xlabel('length') # x축의 이름표
plt.ylabel('weight') # y축의 이름표
plt.show()
# 도미 35마리를 2차원 그래프에 점으로 나타낸다.
# 생선의 길이가 길수록 무게가 많이 나간다. → 선형적
```

```python
# 빙어 데이터

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(smelt_length, smelt_weight) # 차트에 값을 넣는다.
plt.xlabel('length') # x축의 이름표
plt.ylabel('weight') # y축의 이름표
plt.show()
```

```python
# 이 2개의 리스트를 합쳐서 AI 학습용 자료로 만들어야 한다.

length = bream_length + smelt_length # 도미의 길이 리스트 + 빙어의 길이 리스트
weight = bream_weight + smelt_weight # 도미의 무게 리스트 + 빙어의 무게 리스트

# 사이킷런을 활용한 AI 트레이닝 시작 → 2차원 리스트로 만들어야 한다.

print(length)
```

```python
print(weight)
```

image.png

```python
# zip() : 나열된 리스트에서 하나씩 원소를 꺼내어 반환한다.
fish_data = [[l,w] for l, w in zip(length, weight)] # length와 weight를 2차원 리스트로 만든다.
print(fish_data) # 만들어진 2차원 리스트를 출력한다.
```

```python
# 머신러닝은 정답을 target이라는 값으로 알려줘야 한다.
# target 0을 빙어로, 1을 도미로 처리하려 한다.

# 생선 49마리에 대한 정답 리스트를 만들어야 한다. → 머신러닝에서는 정답 데이터를 2진 분류로 테스트한다. (1 : 도미, 0: 빙어)

fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
# pip install scikit-learn (in PyCharm)
# 전처리 데이터가 완성되었으니 사이킷런 패키리를 활용하여 트레이닝을 해보자.

from sklearn.neighbors import KNeighborsClassifier
# 사이킷런 패키지.이웃 모듈   KNeighborsClassifier (k -  최근접 이웃 알고리즘 → 2진 분류용)
# k : 몇 개의 기본값을 보겠다. (기본값 : 5)

kn = KNeighborsClassifier() # import 처리된 객체를 생성하여 변수에 연결한다.
# kn : k - 최근접 이웃 알고리즘 활용 변수
# kn.fit() : 훈련을 시작한다.
# kn.score() : 평가한 후 점수를 보여준다.
# kn.predict() : 결과 보여준 후 샘플을 넣어서 2진 비교한다.
```

```python
# fish_data, fish_target을 전달하여 도미를 찾기 위한 기준을 학습시킨다. (.fit)
# (트레이닝 데이터 : 훈련 정답 데이터)

kn.fit(fish_data, fish_target) # 훈련 끝
```

```python
# kn이 얼마나 잘 훈련되었는지 평가해보자. → .score()
kn.score(fish_data, fish_target) # 1.0 → 100% (정확도)
# 100%가 나오는 이유는 훈련 데이터와 평가 데이터가 일치하기 때문이다.
# 차후에는 더미데이트를 쪼개서 평가하는 것이 좋다.
```

```python
# k - 최근접 이웃 알고리즘을 파악해보자.
# 어떤 데이터에 대한 답을 구할 때, 주위의 다른 데이터를 확인하고 다수를 차지하는 것을 정답으로 출력한다.
```

image.png

```python
kn.predict([[30, 600]]) # .predict[길이, 무게] : 새로운 데이터의 정답을 예측한다. (2진 분류)
# 결론 : 1 → 도미
```

```python
kn.predict([[8, 10]]) # .predict[길이, 무게] : 새로운 데이터의 정답을 예측한다. (2진 분류)
# 결론 : 0 → 빙어
```

```python
print(kn._y) # 정답 파일 (1: 도미, 0: 빙어)
```

```python
# k - 최근접 이웃 알고리즘을 위해 준비해야 할 일은 데이터를 모두 가지고 있는 것이다.
# 새로운 데이터에 대해 예측할 시, 가장 가까운 직선 거리에 어떤 데이터가 있는지 살펴 보기만 한다.
# 단점 : 데이터가 아주 많은 경우 사용하기 어렵다. (데이터가 크기 때문에 메모리가 많이 필요하고, 직선 거리 계산 시간도 오래 걸린다.)
# 객체를 메모리에 만들고 활용한다. → 재부팅 할 경우 fit 다시 해야 한다.
print(kn._fit_X) # ._fit_X : 내장된 변수로 훈련 데이터를 보관한다.
```

```python
# 결론 : k - 최근접 이웃 알고리즘은 훈련 결과 파일이 존재하지 않는다. (메모리에 로드 → 재부팅 시 사라진다.)
# 새로운 데이터가 들어오면, 가장 가까운 데이터를 메모리에서 참고하여 도미, 빙어를 구분한다.
kn.predict([[5, 15]]) # 길이 5cm, 무게 15kg → 0 (빙어)
```

```python
# 객체 생성 시 'n_neighbors = n'
# 참고할 이웃 데이터 수의 기본값 : 5 → 이웃 객체를 5개 찾아서 결론을 낸다.

# 만약 5개가 아닌 다른 값으로 변경하여 데이터를 훈련시키고 싶으면?

kn49 = KNeighborsClassifier(n_neighbors=49) #kn49 변수 : 참고 데이터를 49개로 한 모델
#      KNeighborsClassifier(n_neighbors=5) : 기본값
# 원본 타깃, 원본 데이터가 49이기 때문에 가장 가까운 데이터 49개를 사용한다.

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target) # 훈련 결과 71%로 좋지 않은 성능을 보인다.
```

```python
# 'n_neighbors=값'中 값을 변경해 가면서 최적의 답을 찾는 게 머신러닝 훈련 개발자의 목표이다.

kn49.predict([[5, 15]]) # 길이 5cm, 무게 15g → 도미
```
