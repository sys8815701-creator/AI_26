```python
# 필터가 4개의 출력을 입력에 놓은 위치에 맞게 2차원으로 배치한다.
# 즉, 왼쪽 위, 오른쪽 위, 왼쪽 아래, 오른쪽 아래 모두 4개의 위치에 해당하는 값을 놓는다. → (4, 4) 크기의 입력을 (2, 2) 크기로 압축한다.
# 특성맵 혹은 피처맵(feature map) : 합성곱 계산을 통해 얻은 출력

# 밀집층에서 여러 개의 뉴런을 사용하듯이, 합성곱 층에서도 여러 개의 필터를 사용한다.
# 여러 개의 필터를 사용하면 만들어진 특성 맵은 순서대로 쌓인다. → 3차원 배열 (2, 2, 3) 크기
```

image.png

```python
# 밀집층에 있는 뉴런의 가중치가 모두 다르듯이, 합성곱 층에 있는 필터의 가중치(커널) 또한 모두 다르다.
# 같은 가중치를 가진 필터를 여러 개 사용할 이유가 없다.

# 결과
# 실제 계산은 밀집층과 동일하게 단순히 입력과 가중치를 곱하지만, 2차원 형태를 유지한다는 점에서 다르다.
# 입력보다 훨씬 작은 크기의 커널을 사용하고, 입력을 왼쪽에서 오른쪽으로, 위에서 아래로 이동하면서 2차원 특성 맵을 만든다.
# 2차원 구조를 그대로 사용하기 때문에, 합성곱 신경망이 이미지 처리 분야에서 뛰어난 성능을 발휘한다.
```

```python
# 케라스 합성곱 층 테스트

# 케라스의 층은 모두 keras, layers 패키지 아래 클래스로 구현되어 있다. (합성곱 층도 마찬가지이다.)
# 특별히 왼쪽에서 오른쪽으로, 위에서 아래로 이동하는 합성곱은 'Conv20' 클래스로 제공한다.

from tensorflow import keras
keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu')
#               필터의 개수 : 10
#                      커널 사이즈 : 3 * 3
#                                          활성화 함수 : 렐루

# 주의 사항
# 특성맵은 활성화 함수를 적용한 후에 사용한다.
# 완전 연결 신경망에서처럼, 합성곱 신경망에서도 종종 활성화 함수를 언급하지 않는다.
# 일반적으로 특성맵은 활성화 함수를 통과한 값을 의미한다.
# 커널의 크기는 하이퍼 파라미터로, '3 * 3' 이나 '5 * 5' 크기가 권장된다.
# 케라스 API의 합성곱 층은 'Dense' 층을 사용했던 자리에 'Conv2D' 층을 넣는다.
```

```python
# 패딩 : 배열의 주위를 가상의 원소로 채운다. (바깥쪽 여백) → 실제 값이 아니기 때문에 0으로 채운다.
```

image.png

```python
# 패딩의 역할 : 커널이 도장을 찍을 횟수를 늘린다. → 계산에 실질적인 영향은 없다.
# 그러나 패딩 하지 않으면, 중앙부와 모서리 픽셀이 합성곱에 참여하는 비율은 크게 차이나게 된다. (4 : 1)

# 세임 패딩(padding='same') : 입력과 특성맵의 크기를 동일하게 만들어주기 위해 입력 주위를 0으로 패딩 한다.

# 밸리드 패딩(padding='valid' (기본값)) : 패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성맵을 만든다. → 특성맵의 크기가 줄어든다.

keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu', padding='same')
```

```python
# 스트라이드(stride) : 상하좌우 한 칸씩이 기본값이지만 두 칸씩도 이동할 수 있는 방법 → 특성맵의 크기가 작아진다.

keras.layers.Conv2D(10, kernel_size=(3, 3), activation='relu', padding='same', strides=1)
```

image.png

```python
# 풀링(pooling) : 합성곱 층에서 만든 특성맵의 가로, 세로의 크기를 줄인다. → 특성맵의 개수는 줄어들지 않는다.
# 특성맵에 커널 없는 필터를 적용하는 것과 유사하다.

# 예를 들어 (2,2,3) 크기의 특성맵에 풀링을 적용해보자.
# 마지막 차원인 개수는 그대로 유지하고 너비와 높이만 줄어들기 때문에 (1,1,3) 크기의 특성 맵이 된다.
```

image.png

```python
# 풀링에는 가중치가 없기 때문에, 도장을 찍은 영역에서 가장 큰 값을 고르거나 (최대 풀링) 평균값을 계산한다. (평균 풀링)
# 풀링은 합성곱 층과 뚜렷하게 구분되기 때문에 풀링 층이라고 부른다.

# 예를 들어 (4,4) 크기의 특성맵이 있다고 가정해보자.
# 여기에 (2,2) (최대) 풀링을 적용하면 절반으로 크기가 줄어든다.
# 최대 풀링은 가장 큰 값을 고르기 때문에, 첫 번째 9, 두 번째 7, 세 번째 8, 네 번째 6을 차례대로 고른다.

# 만약 특성맵이 여러 개라면, 동일한 작업을 반복한다. (e.g. 10개의 특성맵이 있다면 풀링을 거친 특성맵 또한 10개이다.)
```

image.png

```python
# 풀링은 패딩, 가중치가 없고 풀링 크기와 스트라이드가 동일하다.
keras.layers.MaxPooling2D(2) # 대부분 풀링의 크기는 2이다.

# 합성곱 층과 마찬가지로 'strides'와 'padding' 매개변수를 제공한다.
# strides의 기본값 : 풀링의 크기와 같기 때문에 따로 지정하지 않는다.
# padding의 기본값 : valid (패딩 하지 않는다.)

keras.layers.MaxPooling2D(2, strides=2, padding='same') # 최대 풀링 코드

keras.layers.AveragePooling2D(2, strides=2, padding='valid') # 평균 풀링 코드 → 특성맵에 있는 중요한 정보를 평균 내어 희석시킬 수 있다.

# 주의 사항 : 풀링은 가로, 세로 방향으로만 진행한다. (특성맵의 개수는 변하지 않는다.)
```

```python
# 합성곱 신경망에서 필터는 이미지에 있는 특징을 찾는다.
# 처음에는 간단한 특징(e.g. 직선, 곡선)을 찾고, 층이 깊어질수록 다양하고 구체적인 특징을 감지할 수 있도록 필터의 개수를 늘린다.
# 또한, 어떠한 특징이 이미지의 어느 위치에 놓이더라도 쉽게 감지할 수 있도록 너비와 높이 차원을 점점 줄여간다.
```

```python
# 합성곱 신경망을 활용한 이미지 분류 실습

# 패션 MNIST를 활용해보자.
# 전처리 : 데이터 스케일을 '0 ~ 255' 사이에서 '0 ~ 1'로 바꾸고, 훈련 세트와 검증 세트로 나눈다.
# 흑백 이미지의 경우, 채널 차원이 없는 2차원 배열이지만 'Conv20' 층을 사용하기 위해 마지막에 채널 차원을 추가한다. → 'reshape' 매서드를 활용하여 전체 배열의 차원을 그대로 유지하면서 마지막에 추가한다.

# 차이점
# 완전 연결 신경망 : 입력 이미지를 밀집층에 연결하기 위해 일렬로 펼쳐야 한다.
# 합성곱 신경망 : 2차원 이미지를 그대로 사용하기 때문에 펼칠 필요가 없다.
```

```python
import tensorflow as tf

tf.keras.utils.set_random_seed(42) # 실행마다 동일한 결과를 얻기 위해, 케라스에 랜덤 시드를 사용하고 텐서플로 연산을 만든다.

from tensorflow import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = \
    keras.datasets.fashion_mnist.load_data() # 패션 MNIST 데이터를 불러온다.

train_scaled = train_input.reshape(-1, 28, 28, 1) / 255.0
# train_input(48000, 28, 28) → train_scaled(48000, 28, 28, 1) (4차원)
#                                                         흑백

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
# 합성곱 신경망을 생성한다. → 합성곱 층으로 이미지에서 특징을 감지한 후, 밀집층으로 클래스에 따른 분류 확률을 계산한다.

model = keras.Sequential() # Sequential : 클래스의 객체를 생성한다.
model.add(keras.layers.Conv2D(
    32, kernel_size=3, activation='relu', padding='same', input_shape=(28, 28, 1)))
#    .add              Conv2D : 'Conv2D' 층을 'add'로 추가한다.
#   32 : 합성곱 층으로 32개의 필터를 사용한다.
#       kernel_size=3 : 커널의 크기 (3, 3)
#                      activation='relu' : 렐루 활성화 함수를 사용한다.
#                                         padding='same' : 세임 패딩을 사용한다.
#                                                         input_shape=(28, 28, 1) : 앞서 변경한 이미지의 차원을 변환한다.

# 결과 : '28 * 28 * 32'으로 출력한다. → 풀링으로 압축한다. (14 * 14 * 32)
```

image.png

```python
model.add(keras.layers.MaxPooling2D(2)) # 풀링 층을 추가한다. → 최대 풀링 (2:2) 특성맵의 크기가 절반으로 줄어든다.
model.summary()

# 결과 : 합성곱 층에서 32개의 필터를 사용했기 때문에, 특성맵의 깊이가 32, 최대 풀링을 통과한 특성맵의 크기가 (14, 14, 32)가 된다.
```

```python
# 합성곱 층에 세임 패딩을 사용해보자. → 입력의 가로, 세로 크기가 유지된다.
model.add(keras.layers.Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(keras.layers.MaxPooling2D(2)) # 크기를 절반으로 줄인다. → 64개의 필터를 사용했으므로 특성맵의 크기가 (7, 7, 64)가 된다.
model.summary()

# 결과 : 픽셀의 크기는 (7, 7)로 줄어들지만 깊이는 64로 깊어진다. → 깊이가 깊어졌기 때문에 특성을 세세하게 파악할 수 있다.
```

image.png

```python
# 3차원 특성맵을 일렬로 펼쳐보자. → 마지막에 10개의 뉴런을 가진 밀집층에서 확률을 계산한다.
# 특성맵을 일렬로 펼쳐서 바로 출력층에 전달하지 않고, 중간에 Dense 은닉층을 추가한다.
# Flatten → Dense 은닉층 → DropOut → Dense 출력층

model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100, activation='relu'))
#                            100 : 은닉층은 100개의 뉴런을 사용한다.
#                                 activation='relu' : 활성화 함수로서 합성곱과 같은 렐루 함수를 사용한다.
model.add(keras.layers.Dropout(0, 4)) # 드롭아웃을 활용하여 은닉층의 과대 적합을 예방한다.
model.add(keras.layers.Dense(10, activation='softmax'))
#                                activation='softmax' : 패션 MNIST는 클래스를 10개로 분류하는 다중 분류이기 때문에, 활성화 함수로서 소프트맥스 함수를 사용한다.
model.summary()
```

```python
# 합성곱 층과 풀링 층의 효과를 확인해보자.

# 첫 번째 합성곱 층을 통과하면서 특성맵의 깊이는 32가 되었다.
# 두 번째 합성곱 층을 통과하면서 특성맵의 크기는 64가 되었다.
# 반면, 특성맵의 가로, 세로 크기는 첫 번째 풀링층에서 절반으로 줄어들었고, 두 번째 풀링에서 다시 절반으로 줄어들었다.
# 따라서 최종 특성맵의 크기는 (7, 7, 64)이다.

# 모델 파라미터 개수
# 1. 첫 번째 합성곱 층 : 3 * 3 * 1 * 32 + 32 = 320
# 필터 : 32개
# 크기 : (3,3,)
# 깊이 : 1
# 절편 : 필터마다 1

# 2. 두 번째 합성곱 층 : 3 * 3 * 32 * 64 + 64 = 18496
# 필터 : 64개
# 크기 : (3,3,)
# 깊이 : 32
# 절편 : 필터마다 1

# 3. 은닉층 (Flatten 클래스) : 3136 * 100 + 100 = 313700
# Flatten 클래스의 (7, 7, 64) 크기 특성맵을 1차원 배열로 펼치면 (3136,) 크기의 배열이 된다. → 100개의 뉴런과 완전히 연결한다.

# 4. 출력층 : 1010
```

```python
# 층의 구성을 그림으로 확인해보자. → keras.utils.plot_model
keras.utils.plot_model(model)
```

```python
# 입력과 출력 크기까지 표시해보자.
keras.utils.plot_model(model, show_shapes=True)
```

```python
# 모델을 컴파일 하고 훈련한다.
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 아담 최적화로 훈련한다.
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-cnn-model.keras',
                                                save_best_only=True)

# 검증이 2번 정도 상향 시 조기 종료를 한다.
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2,
                                                  restore_best_weights=True)

# EarlyStopping 콜백을 통해 조기 종료를 한다.
history = model.fit(train_scaled, train_target, epochs=20,
                    validation_data=(val_scaled, val_target),
                    callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
# 손실 그래프를 그려서 조기 종료가 잘 실행되었는지 확인해보자.

import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train_loss', 'val_loss', 'train_accuracy', 'val_accuracy'])
plt.show()
plt.show()

# 결과 : 검증 세트에 대한 손실은 점차 감소하다가 정체되고, 훈련 세트에 대한 손실은 점점 낮아진다.
```

```python
# 성능을 평가해보자.
model.evaluate(val_scaled, val_target)

# 결과 : 'fit' 매서드의 출력 中 다섯 번째 출력과 유사하다.
```

```python
# 맨 처음 분류했던 테스트 세트를 통해 합성곱 신경망의 일반화 성능을 파악해보자.
test_scaled = test_input.reshape(-1, 28, 28, 1) / 255.0
model.evaluate(test_scaled, test_target)

# 결과 : 테스트 세트의 점수가 검증 세트의 점수보다 약간 작다.
```

```python
# 첫 번째 샘플의 이미지를 활용해보자. → 크기를 (28, 28, 1)에서 (28, 28)로 변경한다.
plt.imshow(val_scaled[0].reshape(28, 28), cmap='gray_r')
plt.show()

# 결과 : 검증용 리스트의 인덱스 0번에 위치한 샘플의 이미지가 출력된다.
```

```python
# predict : 10개의 클래스에 대한 예측 확률을 출력한다.
preds = model.predict(val_scaled[0:1])
print(preds)

# 결과
# ['티셔츠', '바지', '스웨터', '드레스',
#  '코트', '샌달', '셔츠', '스니커즈',
#  '가방', '앵클 부츠']
```
